```python
import pandas as pd
import altair as alt
import numpy as np
import vegafusion as vf
```

```python
N = 1000
df = pd.DataFrame({'location':np.random.choice(['US','UK','NZ'],N),
                   'category':np.random.choice(['low_cat']+3*['med_cat']+6*['high_cat'],N),
                   'value':np.ones(N)})
df.head(10)
```

```python
vf.enable()
# vf.disable()
```

```python
alt.Chart(df).mark_bar().encode(
    x = 'sum(value):Q',
    y = alt.Y('category:N'),
    color = 'location:N'
).properties(
    width = 80
).facet(
    column = 'location:N'
)
```

```python
alt.Chart(df).mark_bar().encode(
    x = 'sum(value):Q',
    y = alt.Y('category:N').sort(
        alt.EncodingSortField(
            field='value',
            op='sum',
            order='descending'
        )
    ),
    color = 'location:N'
).properties(
    width = 80
).facet(
    column = 'location:N'
)
```

```python
alt.Chart(df).mark_bar().encode(
    x = 'sum(value):Q',
    y = alt.Y('category:N').sort(
        alt.EncodingSortField(
            field='value',
            op='sum',
            order='descending'
        )
    )
).properties(
    width = 80
).facet(
    column = 'location:N'
)
```
