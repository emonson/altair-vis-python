### Flatten list

https://stackoverflow.com/questions/952914/how-do-i-make-a-flat-list-out-of-a-list-of-lists

```python
def flatten_list(ll):
    return [item for sublist in ll for item in sublist]
```

```python
np.random.permutation(flatten_list(zz))
```

### I know I say never to iterate over a DataFrame...


```python
score_lists = []
for ii in range(distT.shape[1]):
    xx = distT.iloc[:,ii]
    yy = list(zip(xx, xx.index))
    zz = [a*[b] for a,b in yy]
    score_lists.append(np.random.permutation(flatten_list(zz)))
```

### Need list of rows

The list of lists will be interpreted by pd.DataFrame() as a list of rows, but we've generated a list of columns

```python
scores_df = pd.DataFrame(np.array(score_lists).T, columns=distT.columns)
scores_df.reset_index()
```

```python
scores_tidy = (scores_df
     .assign(respID = 'r' + scores_df.index.astype('str'))
     .melt(id_vars=['respID'], value_name='response')
     .rename(columns={'Vegetable':'veg','Characteristic':'trait'})
)
scores_tidy
```

```python
scores_tidy.to_csv('data/NatureVegSurveyScores.tsv',
                   sep = '\t',
                   encoding = 'utf-8',
                   index = False)
```
